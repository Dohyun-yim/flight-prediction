## 기온 Temperature (temp)

```python
import glob
# 파일이 있는 디렉토리 및 확장자 설정
directory_path = "/Data_path/"
extension = "xlsx"

# 파일 목록 가져오기
files = glob.glob(f"{directory_path}*.{extension}")

for file_path in files:
    # 파일명에서 연도와 월 추출
    file_name = file_path.split("/")[-1]  # 파일 경로에서 파일명만 추출
    year_month = file_name.split("_")[0]  # 파일명에서 연도와 월 추출
    
    # 새로운 파일명 생성 (기존 '기온'을 'temp'으로 변경)
    new_file_name = 'temp_' + year_month
    
    # 엑셀 파일 읽어오기
    df = pd.read_excel(file_path)

    # 예시로 연도, 월, 새로운 컬럼명, 데이터 출력
    print(f"New Column Name: {new_file_name}")
    
    # 각 파일에 대한 데이터프레임을 독립적인 변수로 저장
    globals()[new_file_name] = df
```

### 날짜 column 수정하기

```python
# for문으로 데이터프레임 처리 - 날짜 형태를 20170101 형태로 바꾸기
for year in range(2017, 2020):
    for month in range(1, 13):
        year_month = f'{year}{month:02d}'

        # 변수명 생성
        temp_var_name = f'temp_{year_month}'

        # 변수가 이미 존재하는 경우에만 처리
        if temp_var_name in globals():
            # '날짜' 열을 수정
            globals()[temp_var_name]['날짜'] = globals()[temp_var_name]['날짜'].apply(lambda x: f'{year_month}_{x:02d}')
            globals()[temp_var_name]['날짜'] = globals()[temp_var_name]['날짜'].str.replace('_', '')
```

### 시간형태 맞추기

```python
# 각 월별 데이터프레임을 저장할 리스트
monthly_data = []

# 2017년부터 2019년까지 반복
for year in range(2017, 2020):
    for month in range(1, 13):
        year_month = f'{year}{month:02d}'  # 연도와 월을 합친 문자열 생성

        # 변수명 생성
        temp_var_name = f'temp_{year_month}'

        # 변수가 이미 존재하는 경우에만 처리
        if temp_var_name in globals():
            # 각 월별 데이터프레임을 melt하여 리스트에 추가
            temp_month = globals()[temp_var_name]  # 전역 변수를 가져오기 위해 globals() 함수 사용
            temp_month_melt = pd.melt(temp_month, id_vars=['날짜'], var_name='시간', value_name='온도')
            temp_month_melt['시간'] = pd.to_timedelta(temp_month_melt['시간'].str[:-1].astype(int) - 1, unit='h')
            temp_month_melt['시간'] = temp_month_melt['시간'].astype(str).str[-8:-3]

            # 생성된 데이터프레임을 리스트에 추가
            monthly_data.append(temp_month_melt)
```

```python
# 리스트에 월별 데이터가 들어가있음
monthly_data[24]
```

### 전체 기온 데이터 만들기

```python
# 리스트에 있는 데이터를 모두 하나의 데이터 프레임으로 만들기
temp_all = pd.concat(monthly_data, ignore_index=True)

# 결과 확인
temp_all
```
